```python
def standardize_text(df, text_field):
   df[text_field] = df[text_field].str.replace(r"[^A-Za-z0-9(),!?’@\'\`\"\_\n]", " ")
   return df
```

```python
def replace_apos(df, text_field):
    df[text_field] = df[text_field].str.replace(r"‚Äô", "\'")
    return df
```

```python
apos_data = replace_apos(fake_data, 'title')
apos_data.to_csv('/content/drive/MyDrive/F-news Generator/Dataset/testapos.csv', index=False)
```
